```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
pd.options.display.max_columns = None
df = pd.read_csv('../Resources/AmesHousing.csv')
### metadata: https://rdrr.io/cran/AmesHousing/man/ames_raw.html
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df.dtypes)
```

```python
# Create category_columns and numeric_columns variables
# category_columns = df.dtypes[df.dtypes == "object"].index.tolist()
# numeric_columns = df.dtypes[df.dtypes != "object"].index.tolist()
numeric_columns = ['Lot Frontage','Lot Area', 'Overall Qual', 'Overall Cond', 'Year Built', 'Year Remod/Add', 'Mas Vnr Area', 
                   'BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF', '1st Flr SF', '2nd Flr SF', 
                   'Low Qual Fin SF', 'Gr Liv Area', 'Bsmt Full Bath', 'Bsmt Half Bath', 'Full Bath', 'Half Bath', 
                   'Bedroom AbvGr', 'Kitchen AbvGr', 'TotRms AbvGrd', 'Fireplaces', 'Garage Yr Blt', 'Garage Cars', 
                   'Garage Area', 'Wood Deck SF', 'Open Porch SF', 'Enclosed Porch', '3Ssn Porch', 'Screen Porch', 'Pool Area', 
                   'Misc Val', 'Mo Sold', 'Yr Sold']
category_columns = ['MS SubClass', 'MS Zoning', 'Street', 'Alley', 'Lot Shape', 'Land Contour', 'Utilities', 'Lot Config', 
                    'Land Slope', 'Neighborhood', 'Condition 1', 'Condition 2', 'Bldg Type', 'House Style', 'Roof Style', 
                    'Roof Matl', 'Exterior 1st', 'Exterior 2nd', 'Mas Vnr Type', 'Mas Vnr Area', 'Exter Qual', 'Exter Cond', 
                    'Foundation', 'Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2', 'Heating', 
                    'Heating QC', 'Central Air', 'Electrical', 'Kitchen Qual', 'Functional', 'Fireplace Qu', 'Garage Type', 
                    'Garage Finish', 'Garage Qual', 'Garage Cond', 'Paved Drive', 'Pool QC', 'Fence', 'Misc Feature', 
                    'Sale Type', 'Sale Condition']
```

```python
numeric_columns
```

```python
# Create dummy variables for the category_columns and merge on the numeric_columns to create an X dataset
X = pd.get_dummies(df, columns=category_columns)
X.head()
```

```python
X.columns.to_list()
```

```python
# Fill in missing values in X with zeroes
X = X.fillna(0)
X.head()
```

```python
X.isnull().sum().values.any()
```

```python
# Create a y series from SalePrice
y = X["SalePrice"]
```

```python
# Split X and y into training and testing sets
X = X.drop("SalePrice", axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Scale X_train and X_test
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a neural network model with keras
nn = tf.keras.models.Sequential()
```

```python
len(X_train_scaled[0])
```

```python
# Add a hidden layer with twice as many neurons as there are inputs. Use 'relu'
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 =  1530
# hidden_nodes_layer2 = 5
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)
```

```python
# add an output layer with a 'linear' activation function.
nn.add(tf.keras.layers.Dense(units=1, activation="linear"))
```

```python
# print a summary of the model
nn.summary()
```

```python
# compile the model using the "adam" optimizer and "mean_squared_error" loss function
nn.compile(loss="mean_absolute_error", optimizer="adam", metrics=["accuracy"])
```

```python
# train the model for 100 epochs
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# predict values for the train and test sets
y_pred = nn.predict(X_test_scaled)
y_pred_train = nn.predict(X_train_scaled)
y_pred_train
```

```python
# score the training predictions with r2_score()
r2_score(y_train, y_pred_train)
```

```python
# score the test predictions with r2_score()
r2_score(y_test, y_pred)
```

```python
# create a deep learning model with two hidden layers. Use the same number of nodes as the neural network model.
new_nn = tf.keras.models.Sequential()
new_hidden_nodes_layer1 = 1530
new_hidden_nodes_layer2 = 100
new_nn.add(tf.keras.layers.Dense(units=new_hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))
new_nn.add(tf.keras.layers.Dense(units=new_hidden_nodes_layer2, activation="relu"))
```

```python
# add a linear output node
new_nn.add(tf.keras.layers.Dense(units=1, activation="linear"))
```

```python
# print the deep learning model summary
new_nn.summary()
```

```python
# compile the model
new_nn.compile(loss="mean_absolute_error", optimizer="adam", metrics=["accuracy"])
```

```python
# train the model for 100 epochs
fit_model = new_nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# predict values for the train and test sets
new_y_pred = new_nn.predict(X_test_scaled)
new_y_pred_train = new_nn.predict(X_train_scaled)
# new_y_pred_train
```

```python
# score the training predictions with r2_score()
r2_score(y_train, new_y_pred_train)
```

```python
# score the test predictions with r2_score(), compare values to the neural network model
r2_score(y_test, new_y_pred)
```
